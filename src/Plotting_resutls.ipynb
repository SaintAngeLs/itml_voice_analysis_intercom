```python
'''The choice of test size and validation size combinations is critical in determining the balance between training, validation, and test datasets for optimal model performance. Here's why we chose specific combinations:

1. Balancing Data Allocation

Test Size:
A test size of 0.3, 0.4, or 0.5 is reasonable because it ensures a substantial amount of data is reserved 
for testing the final model's performance on unseen data.
Larger test sizes (e.g., 0.5) provide more robust evaluation but reduce the training and validation data. 
Smaller test sizes (e.g., 0.3) give the model more data for training while still leaving sufficient test data for evaluation.

Validation Size:
Validation sizes of 0.1, 0.15, and 0.2 offer a good range for assessing the model during training 
without taking away too much data from the training set.
A very small validation size (e.g., 0.1) might not represent the validation set adequately, 
while larger sizes (e.g., 0.25) reduce the amount of data used for training, possibly leading to underfitting.

2. Practical Dataset Constraints

Number of Samples:
Given the dataset contains two classes (allowed and disallowed), and the total sample size for each class is not extremely large, 
test and validation sizes must leave sufficient data for training. 
This ensures the model has enough data to learn meaningful patterns.
Avoiding Overfitting:
By leaving a reasonable portion of data in both test and validation sets, 
we avoid overfitting to the training set while still having adequate data to evaluate generalization.
'''
```

```python

# Plotting Test Loss
plt.figure(figsize=(10, 6))
for test_size in df["test_size"].unique():
    subset = df[df["test_size"] == test_size]
    plt.plot(subset["val_size"], subset["test_loss"], marker='o', label=f"Test Size = {test_size}")

plt.title("Test Loss vs Validation Size for Different Test Sizes")
plt.xlabel("Validation Size")
plt.ylabel("Test Loss")
plt.legend()
plt.grid(True)
plt.show()
```
